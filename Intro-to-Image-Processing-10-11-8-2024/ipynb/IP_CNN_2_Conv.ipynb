```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('/content/drive/MyDrive/Colab Notebooks/Intro to Image Processing/2.4_stick.png', cv.IMREAD_GRAYSCALE)

# Define a kernel (e.g., a simple edge detection kernel)
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

# Perform convolution using cv2.filter2D
convolved_image = cv.filter2D(image, -1, kernel)

# Display the results
plt.figure(figsize=(10, 7))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Convolved Image')
plt.imshow(convolved_image, cmap='gray')
plt.axis('off')

plt.show()
```
