```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('/content/drive/MyDrive/Colab Notebooks/Intro to Image Processing/2.2_hole.png', cv.IMREAD_GRAYSCALE);

kernel = np.ones((17, 21), np.uint8)
erode_img = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

fig=plt.figure()
fig.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')
fig.add_subplot(1,2,2)
plt.imshow(erode_img, cmap='gray')
plt.show()
```

```python
img = cv.imread('/content/drive/MyDrive/Colab Notebooks/Intro to Image Processing/2.4_stick.png', cv.IMREAD_GRAYSCALE);

blurred = cv.GaussianBlur(img, (3, 3), 0)
edged = cv.Canny(blurred, 10, 100)
contours, _ = cv.findContours(edged, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

img_color = cv.imread('/content/drive/MyDrive/Colab Notebooks/Intro to Image Processing/2.4_stick.png', cv.IMREAD_COLOR);
cv.drawContours(img_color, contours, -1, (0, 255, 0), 2)
print("# of object is:", len(contours))

fig2=plt.figure()
fig2.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')
fig2.add_subplot(1,2,2)
plt.imshow(img_color)
plt.show()
```

## Act 2.3

```python
img = cv.imread('/content/drive/MyDrive/Colab Notebooks/Intro to Image Processing/2.4_stick.png', cv.IMREAD_GRAYSCALE);
img_color = cv.imread('/content/drive/MyDrive/Colab Notebooks/Intro to Image Processing/2.4_stick.png', cv.IMREAD_COLOR);

blurred = cv.GaussianBlur(img, (3, 3), 0) # to get rid of textures so that shapes remain
edged = cv.Canny(blurred, 10, 100) # detect edges
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE,(8,8))
open_img = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
contours, _ = cv.findContours(open_img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

cv.drawContours(img_color, contours, -1, (255, 0, 255), 2)
print("# of object is:", len(contours))

fig2=plt.figure()
fig2.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')
fig2.add_subplot(1,2,2)
plt.imshow(img_color)
plt.show()
```
