```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/shopping-data.csv")
dataset.shape
dataset.head()
```

```python
# Retrieve cell values at [rowX:rowY, colX:colY]
data = dataset.iloc[:, 3:5].values
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='average')
cluster.fit_predict(data)
```

```python
plt.scatter(data[:,0],data[:,1], c=cluster.labels_, cmap='rainbow') # [rowX:rowY, colX:colY]
```

```python
from statistics import mean
#                   row                 column name
minV0 = min(dataset[cluster.labels_==0]['Age'])
maxV0 = max(dataset[cluster.labels_==0]['Age'])
avgV0 = mean(dataset[cluster.labels_==0]['Age'])

minV1 = min(dataset[cluster.labels_==1]['Age'])
maxV1 = max(dataset[cluster.labels_==1]['Age'])
avgV1 = mean(dataset[cluster.labels_==1]['Age'])

minV2 = min(dataset[cluster.labels_==2]['Age'])
maxV2 = max(dataset[cluster.labels_==2]['Age'])
avgV2 = mean(dataset[cluster.labels_==2]['Age'])

minV3 = min(dataset[cluster.labels_==3]['Age'])
maxV3 = max(dataset[cluster.labels_==3]['Age'])
avgV3 = mean(dataset[cluster.labels_==3]['Age'])

print("Group 0: min =", '%.2f' % minV0, "max =", '%.2f' % maxV0, "avg =", '%.2f' % avgV0)
print("Group 1: min =", '%.2f' % minV1, "max =", '%.2f' % maxV1, "avg =", '%.2f' % avgV1)
print("Group 2: min =", '%.2f' % minV2, "max =", '%.2f' % maxV2, "avg =", '%.2f' % avgV2)
print("Group 3: min =", '%.2f' % minV3, "max =", '%.2f' % maxV3, "avg =", '%.2f' % avgV3)
```
