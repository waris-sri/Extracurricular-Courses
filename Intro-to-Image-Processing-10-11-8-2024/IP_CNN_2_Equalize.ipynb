```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# "grayscale" means 1 color channel only, i.e., (255,0,0) and (255,255,255)/3 are technically grayscales
img = cv.imread('/content/drive/MyDrive/Colab Notebooks/Intro to Image Processing/2.5_blur.png', cv.IMREAD_GRAYSCALE);
equ_img = cv.equalizeHist(img)

plt.imshow(equ_img, cmap='gray')
plt.show()
```

```python
img = cv.imread('/content/drive/MyDrive/Colab Notebooks/Intro to Image Processing/2.7_blur.png')

yuv_img = cv.cvtColor(img, cv.COLOR_BGR2YUV) # BGR because OpenCV orders RGB differently

yuv_img[:, :, 0] = cv.equalizeHist(yuv_img[:, :, 0]) # (row, col, YUV_index) (Y = 0, U = 1, V = 2)
# yuv_img[:, :, 0] = cv.equalizeHist(yuv_img[:, :, 1]) # (row, col, YUV_index) (Y = 0, U = 1, V = 2)
# yuv_img[:, :, 0] = cv.equalizeHist(yuv_img[:, :, 2]) # (row, col, YUV_index) (Y = 0, U = 1, V = 2)

equalized_img = cv.cvtColor(yuv_img, cv.COLOR_YUV2BGR)

# Display the results
plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Equalized Image')
plt.imshow(cv.cvtColor(equalized_img, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
